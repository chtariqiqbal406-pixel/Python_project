```python
#1. code of calculating any table with for loop
def table():
  table_number=int(input("enter the value of table"))
  table_length=int(input("enter the length"))
  print(f"your table no is,{table_number}")
  for i in range(1,table_length+1):
      print(f"{table_number} x {i} = {table_number*i}")
table()
```

```python
#2.prime number checker
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, int(num/2) + 1):
        if (num % i) == 0:
            print(f"{num} is not a prime number")
            break
    else:
        print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# 5. Reverse number
num = int(input("Enter a number: "))
reversed_num = 0

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10

print(f"The reversed number is: {reversed_num}")
```

```python
# 6. Fibonacci sequence
def fs():
    terms = int(input("Enter the number of terms: "))

    if terms <= 0:
        print("Please enter a positive integer.")
    else:
        a, b = 0, 1
        count = 0
        print("Fibonacci sequence:")
        while count <terms:
            print(a)
            a, b = b, a + b
            count += 1

fs()
```
