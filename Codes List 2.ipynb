Q1: Create a list of 7 numbers. Print only the even numbers from the list.

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even_numbers = []
for n in numbers:
  if n%2==0:
    even_numbers.append(n)
print(f"list of all even numbers: {even_numbers}")
```

Q3: Create a list of 6 numbers. Print the maximum and minimum numbers without using max()
and min() functions.

```python
numbers = [23, 65, 42, 78, 94, 22]
maximum = numbers[0]
minimum = numbers[0]
for i in numbers:
    if i > maximum:
      maximum = i
    if i < minimum:
      minimum =i

print(f"Maximum number in list: {maximum}")
print(f"Minimum number in list: {minimum}")
```

Q4: Create a list [2, 4, 6, 8, 10]. Multiply each number by 3 and store the results in a new
