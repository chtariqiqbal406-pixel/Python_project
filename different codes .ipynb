```python
# code of calculating any table with for loop
def table():
  table_number=int(input("enter the value of table"))
  table_length=int(input("enter the length"))
  print(f"your table no is,{table_number}")
  for i in range(1,table_length+1):
      print(f"{table_number} x {i} = {table_number*i}")
table()
```

```python
# code of calculating any table with while loop function
def table_numbr():
  table_number=int(input("enter the value of table"))
  table_length=int(input("enter the length"))
  print(f"your table no is,{table_number}")
  n=1
  while n < table_length+1:
    print(f"{table_number} x {n} = {table_number*n}")
    n+=2

table_numbr()


```
