2. Write a program to access and print only the Age from the dictionary:

```python
my_dict={"Name":"Ali","Age":16,"Class":"10th"}
print(f"My Age:{my_dict['Age']}")
```

3. Add a new key-value pair &quot;grade&quot;: &quot;A&quot; to the dictionary:

```python
my_dict={"Name":"Ali","Age":16,"Class":"10th"}
my_dict["Grade"]="A"
print(f"My updated Dictionary:{my_dict}")
```

5. Write a program to delete the key &quot;city&quot; from this dictionary:

```python
my_dict={"Name":"Ali","Age":16,"Class":"10th","City":"Bahawalpur"}
del my_dict["City"]
print(f"My Dictionary after deleting city name:{my_dict}")
```

6. Write a program to loop through a dictionary and print all keys and values:

```python
my_dict={"Name":"Ali","Age":16,"Class":"10th","City":"Bahawalpur"}
for key,value in my_dict.items():
    print(f"Key:{key},Value:{value}")
```
