# Uppercase and Lowercase
Q2.Input a string and print it in both uppercase and lowercase.

```python
String=input("Enter the word")
print(f"Upper case of word is {String.upper()}")
print(f"Lower case of word is {String.lower()}")
```

# Palindrome Check
Q7. Check if a string is a palindrome (same forwards and backwards).

```python
String = input("Enter a string: ")
if String == String[::-1]:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```
