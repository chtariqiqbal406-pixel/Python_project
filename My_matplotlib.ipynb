```python
from google.colab import drive
drive.mount("/content/Drive")
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Drive/MyDrive/Titanicdataset.csv")
df.head()
print(df.shape)
duplicate_rows_df = df[df.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_df.shape)
plt.bar(df["Name"].head(),df["Age"].head())
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Titanic Dataset")
plt.legend("Data Set")
plt.show()
```

```python
df = pd.read_csv("/content/Drive/MyDrive/Titanicdataset.csv")
df.head()
plt.hist(df["Age"], color="Green",edgecolor="blue")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age in Titanic Dataset")
plt.legend(["Age"])
plt.show()
```
